## ËøõË°åÊï∞ÊçÆÂàÜÊûê

Ê†πÊçÆÊé®ÂØºÁªìÊûúÔºåcross-Kerr ‰∏é g ÁöÑÂÖ≥Á≥ª‰∏∫
$$
\chi = 2\frac{g^2}{\Delta^2}(E_{C1} + E_{C2})
$$
Âõ†Ê≠§ÔºåËÄ¶ÂêàÂº∫Â∫¶Ôºå
$$
g = \sqrt{\frac{\Delta^2 \chi}{2(E_{C1}+E_{C2})}}
$$

```python
chi = np.array([0.00655, 0.00345, 0.001926, 0.001556, 0.001241, 0.000993, 0.000849])
freq1 = np.array([5575.175374, 5017.427494, 4506.253064, 4218.454408, 3931.408011, 3701.112373, 3470.01159])
freq2 = np.array([7150.898937, 7185.252857, 7155.998904, 7193.471754, 7156.482194, 7161.540990, 7180.07531])
alpha1 = np.array([186, 189, 183.789947, 188.085321, 187.099681, 186.722975, 182.649106])
alpha2 = np.array([188, 190, 188.663954, 190.538009, 188.392400, 188.914612, 189.679339])
```

```python
delta = np.abs(freq1 - freq2)
g_disp = np.sqrt(((delta ** 2) * chi) / (2 * (alpha1 + alpha2)))
```

```python
g_disp
```

```python
g_reson = g_disp / np.sqrt(freq1 * freq2) * np.sqrt(4961 * 4926)
```

```python
g_reson
```
