## ËøõË°åÊï∞ÊçÆÂàÜÊûê

Ê†πÊçÆÊé®ÂØºÁªìÊûúÔºåcross-Kerr ‰∏é g ÁöÑÂÖ≥Á≥ª‰∏∫
$$
\chi = \frac{g^2}{\Delta^2}(E_{C1} + E_{C2})
$$
Âõ†Ê≠§ÔºåËÄ¶ÂêàÂº∫Â∫¶Ôºå
$$
g = \sqrt{\frac{\Delta^2 \chi}{(E_{C1}+E_{C2})}}
$$

```python
# Lj1 = [4nH, 6nH, 8nH]
# Lj2 = 10nH
chi = np.array([0.006203, 0.019946, 0.148057])
freq1 = np.array([7787.649400, 6387.976019, 5555.141673])
freq2 = np.array([4969.621612, 4944.085229, 5020.958704])
alpha1 = np.array([430.422693, 300.410642, 228.187631])
alpha2 = np.array([74.806062, 110.736116, 151.768560])
```

```python
delta = np.abs(freq1 - freq2)
# g_disp = np.sqrt(((delta ** 2) * chi) / (alpha1 + alpha2))
g_disp = np.sqrt(((delta) * (delta) * chi) / (alpha1 + alpha2))
```

```python
g_disp
```

```python
g_reson = g_disp / np.sqrt(freq1 * freq2) * 5000
```

```python
g_reson
```

### ÈîôËØØÂÖ¨ÂºèÈ™åËØÅ

ÈîôËØØÂÖ¨ÂºèÔºå
$$
\chi = \frac{g^2}{\Delta}
$$
Âõ†Ê≠§ÔºåËÄ¶ÂêàÂº∫Â∫¶
$$
g = \sqrt{\chi \Delta}
$$

```python
delta = np.abs(freq1 - freq2)
g_disp_error = np.sqrt(chi * delta)
```

```python
g_disp_error
```

```python
g_reson_error = g_disp_error / np.sqrt(freq1 * freq2) * freq2
```
