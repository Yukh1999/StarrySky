## ËøõË°åÊï∞ÊçÆÂàÜÊûê

Ê†πÊçÆÊé®ÂØºÁªìÊûúÔºåcross-Kerr ‰∏é g ÁöÑÂÖ≥Á≥ª‰∏∫
$$
\chi = 2\frac{g^2}{\Delta^2}(E_{C1} + E_{C2})
$$
Âõ†Ê≠§ÔºåËÄ¶ÂêàÂº∫Â∫¶Ôºå
$$
g = \sqrt{\frac{\Delta^2 \chi}{2(E_{C1}+E_{C2})}}
$$

$$
g = \frac{1}{2} \frac{C_{12}}{\sqrt{(C_{1g}+C_{12})(C_{2g}+C_{12})}}\sqrt{\omega_1 \omega_2}

```python
# ËÄ¶ÂêàÂº∫Â∫¶Èô§ÊéâÈ¢ëÁéáÈ°πÂêéÁöÑÊó†ÈáèÁ∫≤ÈÉ®ÂàÜ
g_effCirc_dimless =  0.5 * C_qc / np.sqrt((C_qg + C_qc) * (C_cg + C_qc))
```
