```python
#Lj=[5, 8, 12, 14]nH
freq1 = np.array([7356.247158, 5833.205901, 4775.055980, 4424.191514, ])
freq2 = np.array([7367.728789, 5842.841051, 4783.214342, 4431.791297,])
p11 = np.array([0.524986, 0.523666, 0.526714, 0.496707, ])
p12 = np.array([0.433975, 0.442450, 0.443738, 0.474724, ])
p21 = np.array([0.433813, 0.442128, 0.443336, 0.475077, ])
p22 = np.array([0.524604, 0.523646, 0.526897, 0.496091, ])
```

```python
# Normal Mode ÊñπÊ≥ïÊ±ÇËÄ¶ÂêàÂº∫Â∫¶
g_nm = np.abs(freq1-freq2)/2
g_nm
```

```python
#EPR ÊñπÊ≥ï
EPR_term1 = p11 ** 2 + p12 ** 2 
EPR_term2 = p21 ** 2 + p22 ** 2
EPR_term3 = p11 * p21 + p12 * p22

EPR_param = (freq1 * freq2) * EPR_term3 / (freq1 ** 2 * EPR_term1 + freq2 ** 2 * EPR_term2)
EPR_param
```

```python
p_eff = EPR_term3 / (EPR_term1 + EPR_term2)
p_eff
```

```python
g_EPR = np.sqrt((freq1 - freq2)**2 * EPR_param)
g_EPR
```
