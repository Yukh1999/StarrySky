## ËøõË°åÊï∞ÊçÆÂàÜÊûê

Ê†πÊçÆÊé®ÂØºÁªìÊûúÔºåcross-Kerr ‰∏é g ÁöÑÂÖ≥Á≥ª‰∏∫
$$
\chi = 2\frac{g^2}{\Delta^2}(E_{C1} + E_{C2})
$$
Âõ†Ê≠§ÔºåËÄ¶ÂêàÂº∫Â∫¶Ôºå
$$
g = \sqrt{\frac{\Delta^2 \chi}{2(E_{C1}+E_{C2})}}
$$

```python
# Coupler È¢ëÁéáÁ©øÊèíÂú®‰∏§‰∏™ Qubit ‰πãÈó¥
chi_qq = np.array([0.001759, 0.003962, 0.003934, 0.003660, 0.001017, 0.000617, 0.000795, 0.000807])
chi_1c = [0.038244, 0.094379, 0.146826, 0.212410, 0.026995, 0.031323, 0.027317, 0.024354]
chi_2c = [0.086916, 0.188777, 0.331411, 0.566387, 0.079471, 0.136518, 0.094391, 0.084397]
alpha1 = np.array([191.425400, 192.808986, 198.744490, 202.616435, 194.377939, 194.406871, 198.283578, 189.811947])
alpha2 = np.array([195.558622, 196.915663, 199.094387, 202.701126, 202.481953, 206.007002, 206.485645, 204.144569])
alphac = np.array([178.175011, 209.730646, 220.702223, 229.997790, 213.199274, 220.227530, 212.880223, 212.607208])
freq1 = np.array([2897.820875, 3567.663009, 3622.966018, 3658.749460, 1302.606260, 1302.712336, 1273.812780, 1246.278111])
freq2 = np.array([5106.770667, 5125.756004, 5155.478206, 5204.387893, 2980.898345, 3689.213339, 3300.052629, 3099.017737])
freqc = np.array([16983.837323, 12338.190542, 10150.044782, 8904.533322, 12452.298661, 11191.066371, 12421.794323, 12418.197577])
```

### ËÆ°ÁÆóQQËÄ¶Âêà: $g_{12}$

```python
# ÁêÜÊÉ≥È¢ëÁéá
freq_ideal_1 = 4961
freq_ideal_2 = 4926
freq_ideal_c = 5977
```

```python
delta_12 = np.abs(freq1 - freq2)
g_disp_12 = np.sqrt(((delta_12 + alpha1)*(delta_12 - alpha2) * chi_qq) / (2 * (alpha1 + alpha2)))
```

```python
g_disp_12
```

```python
g_reson_12 = g_disp_12 / np.sqrt(freq1 * freq2) * np.sqrt(freq_ideal_1 * freq_ideal_2)
```

```python
g_reson_12
```

### ËÆ°ÁÆó $g_{1c}$

```python
delta_1c = np.abs(freq1 - freqc)
g_disp_1c = np.sqrt(((delta_1c ** 2) * chi_1c) / (2 * (alpha1 + alphac)))
```

```python
g_disp_1c
```

```python
g_idealFreq_1c = g_disp_1c / np.sqrt(freq1 * freqc) * np.sqrt(freq_ideal_1 * freq_ideal_c)
```

```python
g_idealFreq_1c
```

### ËÆ°ÁÆó $g_{2c}$

```python
delta_2c = np.abs(freq2 - freqc)
g_disp_2c = np.sqrt(((delta_2c ** 2) * chi_2c) / (2 * (alpha2 + alphac)))
```

```python
g_disp_2c
```

```python
g_idealFreq_2c = g_disp_2c / np.sqrt(freq2 * freqc) * np.sqrt(freq_ideal_2 * freq_ideal_c)
```

```python
g_idealFreq_2c
```

### ËÆ°ÁÆóÊòØÂê¶ÂÖ≥Êñ≠

```python
delta_1c_ideal = freq_ideal_c-freq_ideal_1
delta_2c_ideal = freq_ideal_c-freq_ideal_2
g_eff = g_reson_12 - g_idealFreq_2c*g_idealFreq_1c / 2 * (1/delta_1c_ideal+1/delta_2c_ideal)
```

```python
g_eff
```
