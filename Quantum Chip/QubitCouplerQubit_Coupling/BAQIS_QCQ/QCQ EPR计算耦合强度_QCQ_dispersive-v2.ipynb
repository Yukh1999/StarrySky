## ËøõË°åÊï∞ÊçÆÂàÜÊûê

Ê†πÊçÆÊé®ÂØºÁªìÊûúÔºåcross-Kerr ‰∏é g ÁöÑÂÖ≥Á≥ª‰∏∫
$$
\chi = 2\frac{g^2}{\Delta^2}(E_{C1} + E_{C2})
$$
Âõ†Ê≠§ÔºåËÄ¶ÂêàÂº∫Â∫¶Ôºå
$$
g = \sqrt{\frac{\Delta^2 \chi}{2(E_{C1}+E_{C2})}}
$$

```python
# Coupler È¢ëÁéáÁ©øÊèíÂú®‰∏§‰∏™ Qubit ‰πãÈó¥
Lj1 = [150, 150, 150, 150, 150, 160, 200,'ND']
Lj2 = [50, 50, 50, 40, 60, 60, 100, 'ND']
Ljc = [4, 5, 6, 6, 6, 6, 5, 'ND']
chi_qq = np.array([0.001974, 0.002233, 0.002146, 0.001286, 0.003180, 0.002779, 0.005504, 0.047826])
chi_1c = np.array([0.041481, 0.052880, 0.058564, 0.054916, 0.059962, 0.057651, 0.044654, 0.051203])
chi_2c = np.array([0.095833, 0.115572, 0.136732, 0.163141, 0.116075, 0.118753, 0.070279, 0.070279])
alpha1 = np.array([191.389689, 200.403962, 199.038874, 190.256634, 201.167931, 201.116181, 199.987348, 241.470597])
alpha2 = np.array([203.183235, 204.984889, 205.632520, 202.796459, 205.122321 , 207.091652, 204.223839, 253.615120])
alphac = np.array([230.777329, 231.899999, 236.496022, 228.385303, 234.831537, 235.021188, 231.359490, 246.803464])
freq1 = np.array([1292.559457, 1322.669986, 1318.164374, 1288.702682, 1325.214591, 1282.904909, 1144.095852, 891.481216])
freq2 = np.array([2309.936656, 2320.247209, 2323.977735, 2581.595028, 2118.104857, 2128.267181, 1635.907704, 1396.981099])
freqc = np.array([8914.704578, 7981.857652, 7311.408718, 7284.981651, 7283.250694, 7295.558195, 7939.152422, 7700.645772])
```

### ËÆ°ÁÆóQQËÄ¶Âêà: $g_{12}$

```python
# ÁêÜÊÉ≥È¢ëÁéá
freq_ideal_1 = 5000
freq_ideal_2 = 5000
freq_ideal_c = 6000
```

```python
delta_12 = np.abs(freq1 - freq2)
g_disp_12 = np.sqrt(((delta_12 + alpha1)*(delta_12 - alpha2) * chi_qq) / (2 * (alpha1 + alpha2)))
```

```python
g_disp_12
```

```python
g_reson_12 = g_disp_12 / np.sqrt(freq1 * freq2) * np.sqrt(freq_ideal_1 * freq_ideal_2)
```

```python
g_reson_12
```

### ËÆ°ÁÆó $g_{1c}$

```python
delta_1c = np.abs(freq1 - freqc)
g_disp_1c = np.sqrt(((delta_1c + alpha1) * (delta_1c - alphac) * chi_1c) / (2 * (alpha1 + alphac)))
```

```python
g_disp_1c
```

```python
g_idealFreq_1c = g_disp_1c / np.sqrt(freq1 * freqc) * np.sqrt(freq_ideal_1 * freq_ideal_c)
```

```python
g_idealFreq_1c
```

### ËÆ°ÁÆó $g_{2c}$

```python
delta_2c = np.abs(freq2 - freqc)
g_disp_2c = np.sqrt(((delta_2c + alpha2) * (delta_2c - alphac) * chi_2c) / (2 * (alpha2 + alphac)))
```

```python
g_disp_2c
```

```python
g_idealFreq_2c = g_disp_2c / np.sqrt(freq2 * freqc) * np.sqrt(freq_ideal_2 * freq_ideal_c)
```

```python
g_idealFreq_2c
```
