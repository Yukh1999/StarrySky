## È™åËØÅ $g_{12}$

```python
g12_disp = np.sqrt(chi_q1q2 * (freq2 - freq1))
```

```python
g12_disp
```

```python
g12_reson = g12_disp / np.sqrt(freq1 * freq2) * np.sqrt(freq2 ** 2)
```

```python
g12_reson
```

# È™åËØÅ $g_{qc}$

```python
# g_1c
g1c_disp = np.sqrt(chi_1c * np.abs(freq1 - freqc))
g1c_disp
```

```python
g1c_capTerm = g1c_disp / np.sqrt(freq1 * freqc)
g1c_capTerm
```

```python
# g_2c
g2c_disp = np.sqrt(chi_2c * np.abs(freq2 - freqc))
g2c_disp
```

```python
g2c_capTerm = g2c_disp / np.sqrt(freq2 * freqc)
g2c_capTerm
```
